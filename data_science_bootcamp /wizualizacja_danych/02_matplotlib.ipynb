```python
fig = plt.figure(figsize=(12, 10)) #utw√≥rzenie obiektu typu figure oraz ustawienie rozmiaru obiektu
plt.subplot(221) #okre≈õlenie jakƒÖ chcemy mieƒá siatkƒô 2 kolumny, 2 wiersze, 1 wykres
plt.imshow(mpimg.imread('office.jpg')) #wczytanie 1 obrazu
plt.subplot(222) #durgi wykres
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223) #trzeci wykres
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
plt.axis('Off')
```

```python
#Zautomatyzowanie kodu
fig = plt.figure(figsize=(12, 10)) #utw√≥rzenie obiektu typu figure oraz ustawienie rozmiaru obiektu - to zostaje

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  plt.subplot(220 + idx +1)
  plt.imshow(mpimg.imread(img))
  plt.axis('Off')
```
