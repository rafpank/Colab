#Seaborn

Strona biblioteki: https://seaborn.pydata.org

Dokumentacja: https://seaborn.pydata.org/api.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() #od razu ustawienie stylu
sns.__version__
```

```python
df = sns.load_dataset('tips') #zaimportowane bezposrednio z biblioteki seaborn
df.head(7)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
sns.relplot(data=df, x='total_bill', y='tip') #relacja wartości rachunku do napiwku
```

```python
sns.set(font_scale=1.4) #zmiana rozmiaru etykiet
_ = sns.relplot(data=df, x='total_bill', y='tip', hue='sex') #relacja wartości rachunku do napiwku z podziałem na płeć
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue = 'size') #relacja wartości rachunku do napiwku z podziałem na liczbę osób i wybraniem kolorów dla zmiennej size (hue)
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue = 'size', col='time') #relacja wartości rachunku do napiwku z podziałem na liczbę osób oraz czas jedzenia
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue = 'size', row='time') #to samo co wyżej tylko jedno pod drugim
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue = 'size', col='time', row='smoker')
```

```python
sns.catplot(data=df, x='day', y='total_bill', hue='day') #relacja wartości rachunku do dnia tygodnia
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind='swarm')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind='bar')
```

```python
_ = sns.catplot(data=df, x='day', kind='count')
```
