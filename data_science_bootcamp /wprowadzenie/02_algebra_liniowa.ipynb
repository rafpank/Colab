#Norma wektora, długość wektora w R^2
---
Wektor ma początek i koniec, oraz ma kierunek

v = [v1, v2]

Norma euklidesowa w R^n
v = (pierwiastek) \/v1^2+ v2^2 + ... + vn^2

```python
v1 =np.array([-3,4])
v1
```

pierwiastek z 3^2+4^2 = 5

```python
np.linalg.norm(v1) #MODUŁ 'linalg' - odnosi się do algebry liniowej, a funkcja 'norm()' - policzy nam normę wektora v1
```

```python
v2 = np.array([-10,11,13]) #norma Vn (w tym wypadku V3)
np.linalg.norm(v2)
```

#Mnożenie macierzy
Aby móc pomnożyć dwie macierze liczba kolumn w pierwszej macierzy musi być równa liczbie wierwszy w drugiej macierzy
Mnożenie macierzy nie jest przemienne!

***AB != BA***

```python
X = np.array([[5,3],[3,9]])
y = np.array([[1],[-1]])
print(X, '\n')
print(y, '\n')
X.dot(y) #wynik mnożenia macierzy X: 2x2 z macierzą y: 1x1
```

Pamiętaj że tych macierzy nie da się pomnożyć na odwrót bo liczba kolumn pierwszej macierzy jest 1 a liczba wierszy drugiej macierzy wynosi 2

```python
Z = np.dot(X,y) # to samo mnożenie tylko inaczej zapisane
Z
```

```python
Z.shape #sprawdzamy kształt naszej macierzy i jest 2 na 1
```

Poniżej będziemy mnożyć trochę większą macierz
Możemy to zrobić bo
macierz X ma wybiar 2x**3**
a macież Y ma wymiar **3**x3
co oznacza ze ilość kolumn w pierwszej macierzy = liczbie wierszy w 2 macierzy


```python
X = np.array([[2, -1, 3],
              [3, 1, 0]])
Y = np.array([[2, 1, -1],
              [0, -1, 2],
              [3, 2, 0]])

print(X, '\n')
print(Y)
```

```python
Z = np.dot(X, Y) #mnożenie macierzy element po elemencie
Z
```

#Wyznacznik macierzy


Macierz kwadratowa czyli taka, w której liczba kolumn odpowiada liczbie wierszy
|A| o wymiarze n x n
|A| = det A


```python
A = np.array([[2,4],[-1,3]])
np.linalg.det(A) #wyznacznik macierzy A
round(np.linalg.det(A)) #zaaorkąglony wyznacznik macierzy
```
