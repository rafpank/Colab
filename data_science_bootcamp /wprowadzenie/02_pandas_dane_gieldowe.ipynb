liczenie dziennej zmiany % po cenie close

```python
df.close.shift(1) #przesunięcie kolumny o 1 wiersz w dół
```

```python
df['close_shift'] = df.close.shift(1) #dopisanie przesuniętej o jeden w dół kolumny close (nazwanej close_shift)
```

```python
df.head()
```

```python
df['daily_change'] = df.close/ df.close_shift -1 #obliczenie procentowej zmiany dziennej
df.head()
```

```python
df['daily change'] = (df.close/ df.close_shift -1) #stworzenie błędnej zmiennej
df.head()
```

```python
del df['daily change'] #usuwanie błędnej zmiennej
```
