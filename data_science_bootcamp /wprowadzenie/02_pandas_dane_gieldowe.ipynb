```python
def fetch_financial_data(company='AMZN'):
    """
    Pobiera dane dotyczące notowań spółek na giełdzie według symboli na serwisie stooq.pl - domyślnie dla Amazon.
    """
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('META')
df.info()
```
