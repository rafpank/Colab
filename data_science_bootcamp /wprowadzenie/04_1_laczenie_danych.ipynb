```python
import pandas as pd
```

```python
def fetch_fiancial_data(ticker):
  import pandas_datareader.data as web
  return web.DataReader(name=ticker, data_source='stooq')
```

```python
apple = fetch_fiancial_data('AAPL')
amazon = fetch_fiancial_data('AMZN')
google = fetch_fiancial_data('GOOGL')
tesla = fetch_fiancial_data('TSLA')
```

```python
apple.head()
```

```python
apple.info()
```

```python
tesla.info()
```

```python
apple.columns = [ 'apple_' + col.lower() for col in apple.columns]
amazon.columns = [ 'amazon_' + col.lower() for col in amazon.columns]
google.columns = [ 'google_' + col.lower() for col in google.columns]
tesla.columns = [ 'tesla_' + col.lower() for col in tesla.columns]
```

```python
apple.head()
```

#Konkatenacja danych
czyli łączenie danych

```python
df = pd.concat(objs=[apple, amazon, google, tesla], axis=1)
```

Ustawianie wyświetlania tak żeby pozbyć się zapisu wykładniczego czyli np.(E do -10)

funkcja lambda

```python
pd.set_option('display.float_format', lambda x: f'{x:.2f}') # dzięki temu ustawimy opcje wyświetlania. W tym wypadku wyrównanie do 2 miejsc po przecinku
df.describe().T #od transponuje opis dla lepszej widocznosci
```
