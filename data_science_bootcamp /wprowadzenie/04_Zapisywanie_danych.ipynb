```python
import numpy as np
import pandas as pd
```

```python
def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')
```

```python
df = fetch_financial_data('META')
df
```

#Zapis/Odczyt danych z pliku CSV

```python
df.to_csv('meta.csv')
```

```python
df_dec_23 = df[(df.index.month == 12) & (df.index.year == 2023)]
df_dec_23
```

```python
df_dec_23.to_csv('meta_dec_23.csv') #zapisywanie do pliku wyciÄ™ty miesiÄ…c
```
