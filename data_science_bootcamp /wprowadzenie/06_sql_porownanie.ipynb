```python
import numpy as np
import pandas as pd
pd.__version__
```

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

```python
retail = retail_raw.copy()
retail.info()
```

```python
retail.describe()
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
retail.isnull().sum()
```

```python
retail.Quantity >= 0 #zwraca maskę logiczną (True/False)
```

```python
retail = retail[retail.Quantity >= 0]
```

```python
retail.info()
```

```python
#zmianiamy CustomerID na 'object'
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
retail.info()
```

```python
retail.Quantity*retail.UnitPrice
```

```python
#SELECT CustomerID, avg(Revenue), count(*) FROM retial GROUP BY CustomerID;
retail['Revenue'] = retail.Quantity * retail.UnitPrice #tworzymy kolumne 'Revenue' czyli przychód na poszczególną transakcję
```

```python
retail.head(3) #widac ze została dodana kolumna
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size}) #ponieważ chcemy podać 2 funkcje agregujące używamy metody 'aggregate', do której wstawiamy słownik
```

powyższe zwróciło nam średni przychód z poszczególnych transakcji oraz ile razy nasz klient kupował

```python
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size}).rename(columns={'Revenue': 'Revenue_mean', 'CustomerID': 'Count'}) #zmiana nazw kolumn
```

Do następnego zadania potrzebujemy wyciągnąć dzień miesiąca z InvoiceData, a następnie dla danej daty sumować revenu

```python
retail['InvoiceDateDay'] = retail.InvoiceDate.dt.day #tworzymy nową kolumnę z datą
retail.head()
```

```python
#SELECT InvoiceDateDay, sum(Revenue) FROM retial GROUP BY InvoiceDataDay;
results = retail.groupby('InvoiceDateDay').aggregate({'Revenue': np.sum})
results.head()
```

```python
results.plot(kind='bar')
```
