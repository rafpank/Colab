#Łączenie tabel
metodą konkatenacji (dostępna w bibliotece Pandas) w której podajemy LISTĘ (objs) obiektów (w tym przypadku listę DataFrame) i połączymy je względem osi poziomej (axis = 1)

```python
df_con = pd.concat(objs=[df, eth, bnb, pepe], axis=1)
```
